I.   

**Class**

Class is a user-defined data type which is used to declare variables. A class can have both variables and functions as members.


**Objects**

The class variables are known as objects.


**How to create class and object**

In python a class is created by the keyword class . An object is created using the constructor of the class. This object will then be called the instance of the class.


Example:


```
class Dog:
     
    # attribute
    attr1 = "brown"
    attr2 = "dog"
 
    # A sample method 
    def fun(self):
        print("I'm a", self.attr1)
        print("I'm a", self.attr2)
 

# Object instantiation
Max = Dog()
 
# Accessing class attributes
# and method through objects
print(Max.attr1)
Max.fun()
```

**Access Specifiers**

Access specifiers in Python have an important role to play in securing data from unauthorized access and in preventing it from being exploited.

**Public variables**

Public members can be accessed from outside the class.All members in a Python class are public by default.


**Private variables**


Only member functions(functions defined inside the class declaration) can have access to the private data members and private functions.


**Protected variables** 

Data members of a class that can be accessed within the class and the derived classes.Python's convention to make an instance variable protected is to add a prefix _ (single underscore) to it. This effectively prevents it from being accessed unless it is from within a derived class.


**How to define public,private and protected variables**

Public 


Example:



```
class student:
      
     def stud(self, name, age):
           
           # public data members
           self.Name = name
           self.Age = age
 
     # public member function     
     def displayAge(self):
           
           # accessing public data member
           print("Age: ", self.Age)
 
# creating object of the class
obj = ("A2R", 20)
 
# accessing public data member
print("Name: ", obj.Name)
 
# calling public member function of the class
obj.displayAge()
    
```


Private 



```
class Student:
    
     # private members
     __name = None
     __roll = None
     __branch = None
 
     # constructor
     def __init__(self, name, roll, branch): 
          self.__name = name
          self.__roll = roll
          self.__branch = branch
 
     # private member function 
     def __displayDetails(self):
           
           # accessing private data members
           print("Name: ", self.__name)
           print("Roll: ", self.__roll)
           print("Branch: ", self.__branch)
    
     # public member function
     def accessPrivateFunction(self):
            
           # accessing private member function
           self.__displayDetails() 
 
# creating object   
obj = Student("R2J", 1706256, "Information Technology")
 
# calling public member function of the class
obj.accessPrivateFunction()
```






Protected 



```
class Student:
    
     # protected data members
     _name = None
     _roll = None
     _branch = None
    
     def __init__(self, name, roll, branch): 
          self._name = name
          self._roll = roll
          self._branch = branch
    
     # protected member function  
     def _displayRollAndBranch(self):
 
          # accessing protected data members
          print("Roll: ", self._roll)
          print("Branch: ", self._branch)
 
 
# derived class
class S(Student):
 
       def __init__(self, name, roll, branch):
                Student.__init__(self, name, roll, branch)
         
       # public member function
       def displayDetails(self):
                print("Name: ", self._name)
                self._displayRollAndBranch()
 
# creating objects of the derived class       
obj = S("ABC", 1706256, "Information Technology")
 
# calling public member functions of the class
obj.displayDetails()
```

**Inheritance**

One of the important characteristics a class possess is inheritance.
Inheritance is the mechanism of deriving a new class from an existing class.It provides reusablity of code.
(Derived class tries to inherit some or all the traits of base class(existing class).)

Note : Private members of the base class can never be members of a derived class.

**How to implement inheritance**

Python inheritance syntax

```
class BaseClass:
  Body of base class
class DerivedClass(BaseClass):
  Body of derived class
```



Program to depict python inheritance

```
class Person(object):
  
    def __init__(self, name):
        self.name = name
  
    # To get name
    def getName(self):
        return self.name
  
    # To check if this person is an employee
    def isEmployee(self):
        return False
  
  
# Inherited or Derived class 
class Employee(Person):
    def isEmployee(self):
        return True
  
emp = Person("John")  # An Object of Person
print(emp.getName(), emp.isEmployee())
  
emp = Employee("Sana") # An Object of Employee
print(emp.getName(), emp.isEmployee())
```
